```python
tuscany = pd.read_csv("WineItaly.csv")
lombardy = pd.read_csv("WineItaly_lombardy.csv")
piedmont = pd.read_csv("WineItaly_piedmont.csv")
sicily = pd.read_csv("WineItaly_sicily.csv")
veneto = pd.read_csv("WineItaly_veneto.csv")
sardinia = pd.read_csv("WineItaly_sardinia.csv")
central = pd.read_csv("WineItaly_abruzzo.csv")
```

```python
tuscany['latitude'] = 43.458650
tuscany['longitude'] = 11.138920
lombardy['latitude'] = 45.620900
lombardy['longitude'] = 9.768890
piedmont['latitude'] = 44.610841
piedmont['longitude'] = 7.942845
sicily['latitude'] = 39.256430
sicily['longitude'] = 9.060030
veneto['latitude'] = 45.647670
veneto['longitude'] = 11.866530
sardinia['latitude'] = 39.256430
sardinia['longitude'] = 9.060030
central['latitude'] = 42.341035
central['longitude'] = 14.322353
```

```python
it_crawl = pd.concat([tuscany,lombardy, piedmont, sicily,veneto,sardinia, central])
it_crawl.info()
it_crawl.describe()
```

```python
it_crawl['year'] = it_crawl.title.str.extract('((19|20)\d{2})')[0]
it_crawl = it_crawl.dropna(subset=["year"])
it_crawl['year'] = it_crawl['year'].astype("int")
it_crawl["points"] = it_crawl["points"]*20

it_crawl["o_price"] = it_crawl.o_price.str.replace(",","").astype("float")
it_crawl["s_price"] = it_crawl.s_price.str.replace(",","").astype("float")
```

```python
it_crawl = it_crawl[it_crawl["points"] >0 ]
it_crawl = it_crawl.drop_duplicates()
```

```python
replace_variety = {'Sangiovese':3, 'Bordeaux Red Blends':1, 'Tuscan Blends':1, 'Pinot Gris/Grigio':6,
    'Cabernet Sauvignon':29, 'Merlot':22, 'Other Red Blends':1, 'Other White Blends':4, 'Rosé':24, 'Chardonnay':12,
    'Other Red Wine':1, 'Vernaccia':26, 'Vermentino':21, 'Fruit Wine':167.5,
    'Non-Vintage Sparkling Wine':16, 'Red Sparkling Wine':16, 'Muscat':167.5,
    'Nebbiolo':2, 'Barbera':8, 'Cortese':40.5, 'Arneis':40.5, 'Vermouth':167.5, 'Dolcetto':28,
    'Rosé Sparkling Wine':16, "Nero d'Avola":10, 'Other White Wine':167.5, 'Grillo':32, 'Corvina':7,
    'Garganega':14, 'Grenache':167.5, 'Carignan':167.5,'Montepulciano':15,
}
        
it_crawl["vrank"] = it_crawl["variety"].apply(lambda x :replace_variety[x])
```

```python
it_crawl["latitude"] = it_crawl.latitude.round(5)
it_crawl["longitude"] = it_crawl.longitude.round(5)
```

```python
it_tas_pr = pd.read_csv("it_tas_pr.csv")
```

```python
it_tas_pr["latitude"] = it_tas_pr.latitude.round(5)
it_tas_pr["longitude"] = it_tas_pr.longitude.round(5)
```

```python
crawl_italy = pd.merge(it_crawl, it_tas_pr, on=["year", "latitude", "longitude"], how="left")
crawl_italy = crawl_italy.dropna()
```
