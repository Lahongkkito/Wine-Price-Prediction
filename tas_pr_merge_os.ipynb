```python
files = os.listdir("US")

pr_files = [file for file in files if file.split("_")[0] == "pr"]

prs = []

for file in pr_files:
    file_read = pd.read_csv('US/'+"{}".format(file), names=["pr", "year", "month", "lng", "lat"])
    prs.append(file_read)
    

pr_pivot=[]

def make_pr(pr):
    pr  = pr.drop([0])
    pr_pv = pr.pivot("year", "month", "pr")
    pr_pv["lat"] =  float(pr.loc[1].lat)
    pr_pv["lng"] =  float(pr.loc[1].lng)
    return pr_pv

for pr in prs:
    pr_pv = make_pr(pr)
    pr_pivot.append(pr_pv)

us_pr = pd.concat(pr_pivot)

for column in us_pr:
    us_pr.rename(columns = {column : "pr_" + str(column.strip())}, inplace = True)

us_pr.tail()
```

```python
files = os.listdir("US")

tas_files = [file for file in files if file.split("_")[0] == "tas"]

tass = []

for file in tas_files:
    file_read = pd.read_csv('US/'+"{}".format(file), names=["tas", "year", "month", "lng", "lat"])
    tass.append(file_read)
    

tas_pivot=[]

def make_tas(tas):
    tas  = tas.drop([0])
    tas_pv = tas.pivot("year", "month", "tas")
    tas_pv["lat"] =  float(tas.loc[1].lat)
    tas_pv["lng"] =  float(tas.loc[1].lng)
    return tas_pv

for tas in tass:
    tas_pv = make_tas(tas)
    tas_pivot.append(tas_pv)

us_tas = pd.concat(tas_pivot)

for column in us_tas:
    us_tas.rename(columns = {column : "tas_" + str(column.strip())}, inplace = True)

us_tas.tail()
```

```python
us_tas_pr = pd.merge(us_tas, us_pr, left_on=["year", "tas_lat", "tas_lng"], right_on=["year", "pr_lat", "pr_lng"])

us_tas_pr.drop(columns=['tas_lat', 'tas_lng'], inplace=True)
us_tas_pr = us_tas_pr.rename(columns = {"pr_lat": "latitude", "pr_lng":"longitude"})

us_tas_pr = us_tas_pr.drop_duplicates()
us_tas_pr
```
